**EDA multivariante gráfico (Pairs Plots)**

---

Una vez que se tenga un buen conjunto de datos limpios, el siguiente paso será el Análisis de datos exploratorios (EDA). EDA es el proceso de averiguar qué nos pueden decir los datos y usamos EDA para encontrar patrones, relaciones o anomalías para informar nuestro análisis posterior. 

Existe un número casi abrumador de métodos para usar en EDA, sin embargo, una de las herramientas de inicio más efectivas es la gráfica de pares (*Pairs plots*) (también llamada matriz de diagrama de dispersión). El diagrama de pares nos permite ver tanto la distribución de variables individuales como las relaciones entre dos variables. De esta manera se puede identificar las relaciones entre variables.

**1. Importar librerías y dataset**

Las siguientes líneas important la librería seaborn (para visualización), cuya documentación se encuentra en: https://seaborn.pydata.org/ 

Seaborn contiene algunos set de datos para realizar pruebas, por ejemplo: **Iris Data Set** (https://archive.ics.uci.edu/ml/datasets/iris).

El conjunto de datos de la flor de Iris es un **conjunto de datos multivariado** introducido por el estadístico y biólogo británico Ronald Fisher en su artículo de 1936. El uso de mediciones múltiples en problemas taxonómico. Los datos sirven para cuantificar la variación morfológica de las flores Iris de **tres especies relacionadas**. El conjunto de datos consta de 50 muestras de cada una de las tres especies de Iris (Iris Setosa, Iris virginica e Iris versicolor). Se midieron cuatro características de cada muestra: la longitud y el ancho de los sépalos y pétalos, en centímetros.


```

import seaborn as sns; sns.set(style="ticks", color_codes=True)
import pandas as pd
iris = sns.load_dataset("iris")
print(type(iris))
# Mostrando los primeros 5 registros del dataset
iris.head(5)
```

```
# Mostrando los últimos 5 registros del dataset
iris.tail(5)
```

**2. Graficar Pair plot del dataset**

Seaborn tiene una función que realiza de manera directa esta visualización (no se utiliza la variable categórica especies, solamente los valores numéricos existentes en el DataFrame):

```
g = sns.pairplot(iris)
```

**hue : string (variable name), optional**

    Variable in data to map plot aspects to different colors
    
*  La pdf del ancho y largo de los petalos y sepalos son caracteristicas suficientes para diferenciar la setosa de las demás. Color azul en los scatters.

```
g = sns.pairplot(iris, hue="species")
```

Se puede filtrar variables de interes

```
g = sns.pairplot(iris, vars=["petal_width", "petal_length"], hue="species")
```

**diag_kind** : {‘auto’, ‘hist’, ‘kde’}

Kind of plot for the diagonal subplots. The default depends on whether "hue" is used or not.

**Kde (kernel density estimation):** es una forma no paramétrica de estimar la función de densidad de probabilidad de una variable aleatoria. La estimación de la densidad del núcleo es un problema fundamental de suavización de datos en el que se hacen inferencias sobre la población, basadas en una muestra de datos finitos.

El siguiente link explica de manera dinámica lo que es KDE: https://mathisonian.github.io/kde/


```
g = sns.pairplot(iris, diag_kind="hist", vars=["petal_width", "petal_length"])
```

```
g = sns.pairplot(iris, diag_kind="kde", vars=["petal_width", "petal_length"])
```

Los datos presentan algun tipo de correlación?

**kind : {‘scatter’, ‘reg’}, optional**

    Kind of plot for the non-identity relationships

```
g = sns.pairplot(iris, kind="reg", vars=["sepal_width", "petal_width", "petal_length"])
```
